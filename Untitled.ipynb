```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_theme()
```

```python
df = pd.read_excel("ShopSatisfactionSurveyData.xlsx")
```

```python
df.info()
```

```python
df.head()
```

```python
df.columns = ["H", "gender", "age", "family_size", "house", "education", "net_h_income", "h_spending", "range_sat", "price_sat"]
```

```python
df.set_index("H", inplace=True)
```

```python
sns.pairplot(data = df, hue="gender")
```

```python
sns.scatterplot(data = df, x = "net_h_income", y = "price_sat", hue = "gender")
```

```python
avg_income = sum(df["net_h_income"].dropna())/len(list(df["net_h_income"].dropna()))
```

```python
avg_income
```

```python
below_avg_inc = df[df["net_h_income"] < avg_income]["net_h_income"]
```

```python
above_avg_inc = df[df["net_h_income"] > avg_income]["net_h_income"]
```

```python
len(list(below_avg_inc))
```

```python
len(list(above_avg_inc))
```
